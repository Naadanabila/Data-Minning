# Apa saja langkah-langkah untuk membuat alur kerja analisis data secara efektif?

1 Definisikan masalahnya

    Langkah pertama dari setiap alur kerja analisis data adalah menentukan masalah atau pertanyaan yang ingin Anda jawab dengan data tersebut. Hal ini akan membantu Anda mempersempit ruang lingkup, tujuan, dan hasil yang diharapkan dari analisis, serta kriteria keberhasilan dan evaluasi. Anda juga harus mengidentifikasi pemangku kepentingan, audiens, dan konteks permasalahan, serta bagaimana hasil analisis akan dikomunikasikan dan digunakan. Masalah yang terdefinisi dengan baik akan membantu Anda memfokuskan upaya dan menghindari pemrosesan data yang tidak perlu atau tidak relevan.

2 Kumpulkan dan bersihkan datanya

    Langkah selanjutnya adalah mengumpulkan dan membersihkan data yang relevan dengan masalah. Pengumpulan data melibatkan pencarian, akses, dan penggalian data dari berbagai sumber, seperti database, API, halaman web, file, atau survei. Pembersihan data melibatkan pemeriksaan, koreksi, dan penghapusan kesalahan, ketidakkonsistenan, atau nilai yang hilang dalam data, serta pemformatan dan penataan data dengan cara yang konsisten dan nyaman. Pengumpulan dan pembersihan data merupakan langkah penting untuk memastikan kualitas, keandalan, dan validitas data untuk analisis.

3 Jelajahi dan analisis data

    Setelah Anda memiliki kumpulan data yang bersih dan terorganisir dengan baik, Anda dapat mulai menjelajahi dan menganalisis data untuk menemukan pola, tren, hubungan, dan outlier. Eksplorasi dan analisis data melibatkan penerapan berbagai teknik statistik dan matematika, seperti statistik deskriptif, pengujian hipotesis, korelasi, regresi, pengelompokan, klasifikasi, dan reduksi dimensi, untuk merangkum, membandingkan, dan menyimpulkan informasi dari data. Anda juga dapat menggunakan berbagai alat visualisasi data, seperti bagan, grafik, peta, dan dasbor, untuk menyajikan dan menjelajahi data secara visual dan interaktif.

4 Interpretasikan dan komunikasikan hasilnya

    Langkah terakhir dari alur kerja analisis data adalah menafsirkan dan mengkomunikasikan hasil analisis kepada pemangku kepentingan atau audiens. Hal ini mencakup penjelasan temuan utama, wawasan, dan implikasi data secara jelas dan ringkas, serta pemberian rekomendasi, solusi, atau tindakan berdasarkan hasil tersebut. Anda juga harus mengatasi segala keterbatasan, asumsi, atau ketidakpastian analisis, dan memberikan bukti dan referensi untuk mendukung klaim Anda. Anda dapat menggunakan berbagai format dan saluran untuk mengkomunikasikan hasilnya, seperti laporan, presentasi, kesimpulan

b. Apakah ada perbedaan harga jual antara merek mobil tertentu?

```python
merek_maruti_data = df[df["merek"]=="Maruti"]
merek_Hyundai_data = df[df["merek"]=="Hyundai"]
merek_Mahindra_data = df[df["merek"]=="Mahindra"]
merek_Tata_data = df[df["merek"]=="Tata"]
merek_Honda_data = df[df["merek"]=="Honda"]
```

```python
a = merek_maruti_data.groupby("year")["selling_price"].sum()
b = merek_Hyundai_data.groupby("year")["selling_price"].sum()
c = merek_Mahindra_data.groupby("year")["selling_price"].sum()
d = merek_Tata_data.groupby("year")["selling_price"].sum()
e = merek_Honda_data.groupby("year")["selling_price"].sum()
```

```python
a = pd.DataFrame(a)
b = pd.DataFrame(b)
c = pd.DataFrame(c)
d = pd.DataFrame(d)
e = pd.DataFrame(e)
```

```python
a.reset_index(inplace=True)
b.reset_index(inplace=True)
c.reset_index(inplace=True)
d.reset_index(inplace=True)
e.reset_index(inplace=True)
```

```python
a["merek"] = "Maruti"
b["merek"] = "Hyundai"
c["merek"] = "Mahindra"
d["merek"] = "Tata"
e["merek"] = "Honda"
```

```python
gabungan = pd.concat([a,b,c,d,e],axis=0)
gabungan
```

```python
df_gabungan_line2 = gabungan.pivot(index='year', columns='merek', values='selling_price').reset_index()
df_gabungan_line2.fillna(0,inplace=True)
df_gabungan_line2
```

```python
# Membuat grafik garis
plt.figure(figsize=(10, 6))
plt.plot(df_gabungan_line2['year'], df_gabungan_line2['Hyundai'], label='Hyundai', marker='*') #titiknya jadi bintang
plt.plot(df_gabungan_line2['year'], df_gabungan_line2['Maruti'], label='Maruti', marker='o') #titiknya jadi lingkaran
plt.plot(df_gabungan_line2['year'], df_gabungan_line2['Mahindra'], label='Maruti', marker='x') #titiknya jadi x
plt.plot(df_gabungan_line2["year"],df_gabungan_line2["Tata"],label="Tata",marker="D") # titiknya jadi diamond
plt.plot(df_gabungan_line2["year"],df_gabungan_line2["Honda"],label = "Honda",marker="^") #titiknya jadi segitiga

# Menambahkan judul dan label sumbu
plt.title('Total Harga Jual Historis Mobil')
plt.xlabel('Tahun')
plt.ylabel('Harga Mobil')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.grid(True)
plt.show()
```
